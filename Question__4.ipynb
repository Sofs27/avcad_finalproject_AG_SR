Import Dataset

```python
import pandas as pd

# Load the dataset
file_path = "Registos_Incendios_SGIF_2011_2020.xlsx"
data = pd.read_excel(file_path)

# Display the first few rows of the dataset and check columns
print(data.head())
```

Which district has the most fire occurrences? What’s the number of fires of the district that was more affected?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count fires per district
fires_per_district = data['Distrito'].value_counts().reset_index()
fires_per_district.columns = ['Distrito', 'Num_Fires']
fires_per_district = fires_per_district.sort_values(by='Num_Fires', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=fires_per_district, x='Distrito', y='Num_Fires', palette='Oranges_r')
plt.xticks(rotation=90)
plt.xlabel('District')
plt.ylabel('Number of Fires')
plt.title('Distribution of Fires per District (2011-2020)')
plt.tight_layout()
plt.show()

# Print the district with the most fires
district_most_fires = fires_per_district.iloc[0]['Distrito']
num_most_fires = fires_per_district.iloc[0]['Num_Fires']
print(f"The district with the most fires is {district_most_fires} with {num_most_fires} fires.")
```
